```python
from bs4 import BeautifulSoup
soup = BeautifulSoup('<a><\p>', 'html.parser')
print('html.parser')
print(soup)
print('-'*40)
# <a>&lt;\p&gt;</a>
#올바른 markup으로 변경

soup = BeautifulSoup('<a><\p>', 'lxml')
print('lxml')
print(soup)
print('-'*40)
# <html><body><a>&lt;\p&gt;</a></body></html>
#전체 돔트리를 재현

soup = BeautifulSoup('<a><\p>', 'xml')
print('xml')
print(soup)
print('-'*40)
#싱글 태그 <a/> --> 열고 닫고를 한번에 처리

soup = BeautifulSoup('<a><\p>', 'html5lib')
print('html5lib')
print(soup)
print('-'*40)
#<html><head></head><body><a>&lt;\p&gt;</a></body></html>
#돔트리를 재현하지만 부가적인 태그를 더 만들어냄
#처리속도가 가장 느림
```

# 기본 사용법

```python
import requests 
from bs4 import BeautifulSoup

url = 'https://ko.wikipedia.org/wiki/%EC%9B%B9_%ED%81%AC%EB%A1%A4%EB%9F%AC'
# 한글이 유니코드로 관리됨
response = requests.get(url)

#soup 객체 생성
soup = BeautifulSoup(response.text, "html.parser") 

#태그를 이용한 접근
print(1, soup.title) #타이틀 이라는 태그 --> 가장 먼저 있는 태그를 찾아줌
print('-'*10)
print(2, soup.footer.ul.li) 
print('-'*10)
#태그의 조합으로 불러옴
#footer --> 저작권, 회사 정보 등등이 들어있음
#li = 리스트 태그
print(3, soup.footer.ul.li.text) 
print('-'*10)
#뒤에 text 를 쓰면 태그 빼고 내용만 가져와
print(4, soup.a)
print('-'*10)
# soup 객체에서 첫번째로 있는 a element
# a 태그는 하이퍼링크이다.


#태그와 속성을 이용한 접근 : 태그이름[속성 이름]
# 속성이 존재하지 않으면 err 를 발생시킴
#print(soup.a['id'])
print(5, soup.li['id'])

print('-'*10)
#find 함수를 이용해서 이용한 정보를 추출
# a element 속성 중 title 속성의 값이 '구글봇'인 데이터를 검색
soup.find('a', attrs={'title':'구글봇'})

#title 속성의 값이 '구글봇'인 어떤 element를 찾아 출력
soup.find(attrs={'title':'구글봇'})
```

```python
import requests
from bs4 import BeautifulSoup
url = ('https://movie.daum.net/ranking/reservation')
response = requests.get(url)
response.raise_for_status()
soup = BeautifulSoup(response.text, 'lxml')
poster_img = soup.find_all('img', attrs={'class':'img_thumb'})
print(poster_img)
#. -->  현재경로  ..--> 상위경로
#이미지를 저장할 폴더 생성

import os
img_dir = 'c:/Temp/poster_img/'

if not os.path.exists(img_dir): #현재 경로에 poster_img 폴더가 존재하지 않으면
    os.makedirs(img_dir)
    print('폴더 생성 완료')
else:
    print('폴더가 존재함')

for i,movie in enumerate(poster_img,1):
    print(movie)
    title = movie.get('alt')
    img_url = movie.get('src')
    print(i,':', img_url)
    img_res = requests.get(img_url)
    
    if ':' in title:
        title = title.replace(':',' ')# 영화제목의 특수문자 제거
    
    with open(img_dir+'m{0:02d}_{1}'.format(i,title)+'.jpg','wb')as f:  
        f.write(img_res.content) # binary 데이터니까 


print('저장 완료')
```

# [실습] 네이버 경제 관련 언론사별 랭킹 뉴스 추출하기

```python
import requests
from bs4 import BeautifulSoup

url = 'https://news.naver.com/main/ranking/popularDay.naver'
headers = {'user-agent':'Mozilla/5.0 (Windows NT 11.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.5414.120 Safari/537.36 NetHelper70'}
response = requests.get(url, headers = headers)
response.raise_for_status()

soup = BeautifulSoup(response.text, 'lxml')
#temp = soup.find_all('div', attrs= {'class': '_officeCard _officeCard0'})
#print(len(temp))

temp2 = soup.find_all('div', attrs={'class':'rankingnews_box'})
print(len(temp2))
print(temp2[0].strong.get_text())

for i in range(len(temp2)):
    title = temp2[i].strong.get_text() #언론사 이름 추출
    #경제 관련 언론사 추출
    if '경제' not in title:
        continue
    else: 
        print('-'*15)
        print(f'[{title}]')
    for j in range(len(temp2[i])):
        subject = temp2[i].find_all('a', attrs= {'class':'list_title'})[j].get_text()
        print(f'{j+1} : {subject}')
```
