1. 네이버 날씨 정보 스크래핑 하기 (30점)
<br>
• 다음 URL을 이용하여 네이버 날씨 정보를 스크래핑 한 후 제시된 UI 형태의 어플리케이션을 작성하시오.
<br>
• URL : https://search.naver.com/search.naver?query=날씨
<br>
• 어플리케이션은 텍스트 기반의 메뉴로 구성되어 있으며 종료 메뉴를 선택하기 전까지 반복 수행된다.
<br>
• 접속 시간은 datetime 패키지를 이용한다.

```python
import requests 
from bs4 import BeautifulSoup
import datetime
from datetime import datetime

url = 'https://search.naver.com/search.naver?query=날씨'
response = requests.get(url)
response.raise_for_status()

# 현재시각
time = datetime.now()
print(time)
# 장소
soup = BeautifulSoup(response.text, "html.parser") 

find_location = soup.select_one('#main_pack > section.sc_new.cs_weather_new._cs_weather > div._tab_flicking > div.top_wrap > div.title_area._area_panel > h2.title')
location = find_location.text


find_situation = soup.select_one('#main_pack > section.sc_new.cs_weather_new._cs_weather > div._tab_flicking > div.content_wrap > div.open > div > div > div.weather_info > div > div._today > div.weather_graphic > div.weather_main > i > span') 
situation = find_situation.text


find_temperature = soup.select_one('#main_pack > section.sc_new.cs_weather_new._cs_weather > div._tab_flicking > div.content_wrap > div.open > div > div > div.weather_info > div > div._today > div.weather_graphic > div.temperature_text > strong')
temperature = find_temperature.text


find_lowest_temp = soup.select_one('#main_pack > section.sc_new.cs_weather_new._cs_weather > div._tab_flicking > div.content_wrap > div.content_area > div.inner > div > div.list_box._weekly_weather > ul > li > div > div.cell_temperature > span > span.lowest')
lowest_temp = find_lowest_temp.text


find_highest_temp = soup.select_one('#main_pack > section.sc_new.cs_weather_new._cs_weather > div._tab_flicking > div.content_wrap > div.content_area > div.inner > div > div.list_box._weekly_weather > ul > li > div > div.cell_temperature > span > span.highest')
highest_temp = find_highest_temp.text


find_fine_dust = soup.select_one('#main_pack > section.sc_new.cs_weather_new._cs_weather > div._tab_flicking > div.content_wrap > div.open > div > div > div.weather_info > div > div.report_card_wrap > ul > li > a > span')
fine_dust = find_fine_dust.text


while True:
    print()
    print(f"""==============================
날씨 확인 프로그램
==============================
접속 시간: {time}
접속 위치: {location}
------------------------------
1. 현재 날씨 및 온도 확인
2. 최저/최고 온도 확인
3. 미세먼지 확인
0. 종료
------------------------------""")
    
    num = int(input('번호를 입력하세요.:'))
    print('==============================')
    
    if num == 0:
        print('날씨 확인 프로그램을 종료합니다')
        break
    
    elif num == 1:
        print('현재 날씨 확인')
        print('==============================')
        print(f"""------------------------------
{situation} / {temperature}
------------------------------""")
        
    elif num == 2:
        print('최저/최고 온도 확인')
        print('==============================')
        print(f"""------------------------------
{lowest_temp}
{highest_temp}
------------------------------""")
    
    elif num == 3:
        print('미세먼지 확인')
        print('==============================')
        print(f"""------------------------------
{fine_dust}
------------------------------""")
        
    else:
        print('없는 메뉴입니다. 올바른 번호를 선택해 주세요')
    
```

2. 네이버에서 bts이미지를 검색한 후 나온 결과 중 5개의 이미지를 저장하기 (25점)
• 다음의 처리 단계는 Selenium을 이용한다.
(1) 네이버 홈페이지 요청
(2) 검색 필드에 ‘bts’ 검색어 입력 후 검색
(3) 이미지 검색결과 선택
(4) 최초 검색된 bts 이미지 5장 저장

```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import os
import requests

PAUSE_TIME = 1 
SEARCH_KEYWORD='bts'

driver=webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=webdriver.ChromeOptions())

driver.get('https://www.naver.com/')
time.sleep(PAUSE_TIME)

element = driver.find_element(by = By.CSS_SELECTOR, value = '#query')
element.send_keys(SEARCH_KEYWORD)
element.send_keys(Keys.ENTER)
time.sleep(PAUSE_TIME)

element_image = driver.find_element(by = By.CSS_SELECTOR, value = '#lnb > div.lnb_group > div > ul > li:nth-child(2) > a')
element_image.click()
time.sleep(PAUSE_TIME)

image_list = driver.find_elements(by = By.CSS_SELECTOR, value ='#main_pack > section.sc_new.sp_nimage._prs_img._imageSearchPC > div > div.photo_group._listGrid > div.photo_tile._grid > div > div > div.thumb > a > img')

directory = 'c:/Temp/'+SEARCH_KEYWORD+'/'
if not os.path.exists(directory):
    os.mkdir(directory) 
    print('파일을 만들었습니다')
else:
    print('파일이 이미 존재합니다.')
    
img_cnt = 1
for image in image_list:
    link = image.get_attribute('src')
    
    if not link.startswith('http'):
        continue
        
    link_response = requests.get(link)
    with open(directory+SEARCH_KEYWORD+str(img_cnt)+'.jpg', 'wb') as f:
        f.write(link_response.content)   
        
    img_cnt += 1
            
    if img_cnt == 6:
        break

print('저장완료')
```

```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import os
import requests

PAUSE_TIME = 1 
SEARCH_KEYWORD='bts'

driver=webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=webdriver.ChromeOptions())

driver.get('https://www.naver.com/')
time.sleep(PAUSE_TIME)

element = driver.find_element(by = By.CSS_SELECTOR, value = '#query')
element.send_keys(SEARCH_KEYWORD)
element.send_keys(Keys.ENTER)
time.sleep(PAUSE_TIME)

element_image = driver.find_element(by = By.CSS_SELECTOR, value = '#lnb > div.lnb_group > div > ul > li:nth-child(2) > a')
element_image.click()
time.sleep(PAUSE_TIME)

image_list = driver.find_elements(by = By.CSS_SELECTOR, value ='#main_pack > section.sc_new.sp_nimage._prs_img._imageSearchPC > div > div.photo_group._listGrid > div.photo_tile._grid > div > div > div.thumb > a > img')

directory = 'c:/Temp/'+SEARCH_KEYWORD+'/'
if not os.path.exists(directory):
    os.mkdir(directory) 
    print('파일을 만들었습니다')
else:
    print('파일이 이미 존재합니다.')
    
img_cnt = 0
for image in image_list:
    link = image.get_attribute('src')
    
    if not link.startswith('http'):
        continue
    img_cnt += 1
    print(f'[{img_cnt}], {link}')
    ext =''
    extension = link.split('.')[-1]
    if extension in ['jpg','JPG','jpeg','JPEG','png','PNG','gif','GIF']:
        ext = '.'+extension
    else:
        ext = '.jpg'
    link_response = requests.get(link)
    with open(directory+SEARCH_KEYWORD+str(img_cnt)+ext, 'wb') as f:
        f.write(link_response.content)   
            
    if img_cnt == 5:
        break

print('저장완료')
```

3. 멜론 사이트 최신곡 검색 (25점)
- 다음 URL을 이용하여 멜론 사이트에서 국내 최신곡 10위까지의 정보(노래제목, 가수, 앨범)를 출력하시오.
- URL : https://www.melon.com/new/index.htm

```python
import requests 
from bs4 import BeautifulSoup
import datetime
from datetime import datetime
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

URL = 'https://www.melon.com/new/index.htm'
PAUSE_TIME = 1 
driver=webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=webdriver.ChromeOptions())

driver.get(URL)
time.sleep(PAUSE_TIME)

element = driver.find_element(by = By.CSS_SELECTOR, value = '#gnb_menu > ul:nth-child(1) > li.nth1 > a > span.menu_bg.menu01')
element.click()
time.sleep(PAUSE_TIME)

soup = BeautifulSoup(driver.page_source, 'html.parser')
songList = soup.select('#lst50')

cnt_song = 0
for song in songList:
    cnt_song +=1
    print(cnt_song)
    find_title = song.select_one('#lst50 > td > div > div > div.ellipsis.rank01 > span > a')
    title = find_title.get('title')
    print('제목:', str(title)[:-2])
    
    find_singer = song.select_one('#lst50 > td > div > div > div.ellipsis.rank02 > a')
    singer = find_singer.get('title')
    print('가수:', str(singer)[:-9])
    
    find_album = song.select_one('#lst50 > td:nth-child(7) > div > div > div')
    album = find_album.text
    print('앨범:', album.strip())
    print('-'*30)
    
    if cnt_song == 10:
        break
```

5. 이메일 주소 추출하기 (10점)
- 다음 문자열에서 정규 표현식을 이용하여 이메일 주소를 찾아 출력하시오.

- s = 'Ryan has sent an invoice email to john.d@yahoo.com by using his email id
ryan.arjun@gmail.com and he also shared a copy to his boss
rosy.gray@amazon.co.uk on the cc part.'
- 출력결과: ['john.d@yahoo.com', 'ryan.arjun@gmail.com', 'rosy.gray@amazon.co.uk']

compile 방법 1

```python
import re
a = 'Ryan has sent an invoice email to john.d@yahoo.com by using his email id ryan.arjun@gmail.com and he also shared a copy to his boss rosy.gray@amazon.co.uk on the cc part.'
pat = re.compile(r'\b[a-z]+\.[a-z]+\@[a-z]+\.com|\b[a-z]+\.[a-z]+\@[a-z]+\.co.uk')
my_list = pat.findall(a)
print(my_list)
```

compile 방법 2

```python
import re
c = 'Ryan has sent an invoice email to john.d@yahoo.com by using his email id ryan.arjun@gmail.com and he also shared a copy to his boss rosy.gray@amazon.co.uk on the cc part.'
cat = re.compile(r'\b[a-z]+\.[a-z]+\@[a-z]+\.[a-z]+\.*[a-z]+')
my_list = cat.findall(c)
print(my_list)
```
