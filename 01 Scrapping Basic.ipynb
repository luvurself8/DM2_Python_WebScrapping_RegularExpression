- 특정 url에 접속해 결과물을 file로 저장

```python
URL = "http://www.google.com/search"
param = {"q":"python"} #query string에 들어갈 것을 key와 value의 형태로 딕셔너리로 전달 
response = requests.get(URL, param)
#200이 아니면 오류를 내고 프로그램 실행 종료
response.raise_for_status()

print("상태코드:", response.status_code)
result = response.text  #응답메세지에 포함되에있는 html을 익음

with open ("mygoogle.html", "w", encoding = "utf-8") as f:
    f.write(result)
    
print("저장완료")
```

## 로봇이 아님을 나타내기 위해서 user-agent라는 값을 header에 넣어서 보내기
- user-agent 헤더 값을 포함하여 요청하면 브라우저를 통해 요청하는 것으로 인식되어 봇의 접근 차단을 피할 수 있다.
- 브라우저 별 user-agent 값 얻기
-F12를 눌러 개발자 모드 진입 -> Console -> navigator.userAgent 입력

```python
URL = "https://kr.investing.com/economic-calendar/south-korean-fx-reserves---usd-784"
#query string에 들어갈 것을 key와 value의 형태로 딕셔너리로 전달

#로봇이 아님을 나타내기 위해서 user-agent라는 값을 header 에 넣어서 보냄
headers = {"user-agent":'Mozilla/5.0 (Windows NT 11.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.5414.75 Safari/537.36 NetHelper70'}

response = requests.get(URL, param, headers=headers)

result = response.text
#print(result)
```
