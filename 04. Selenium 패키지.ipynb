# 웹 브라우저 자동 스크롤

## 구글에서 이미지 검색 후 검색 결과를 6번 스크롤하기

```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time # time 모듈

# 대문자로 정의하면 상수로 인식
SCROLL_PAUSE_TIME = 1 # 페이지가 로드될 때까지 기다리는 시간
SEARCH_KEYWORD = 'python'

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=webdriver.ChromeOptions())

driver.get("https://www.google.com")
element = driver.find_element(by=By.CLASS_NAME, value='gLFyf')
element.send_keys(SEARCH_KEYWORD)
# element.send_keys(Keys.ENTER)
element.submit() 
# submit(): html의 form 태그의 값을 서버로 전송할 때 호출하는 함수
# form 태그: 사용자로부터 어떠한 값을 입력받을 수 있는 모든 종류의 태그

driver.find_element(by=By.LINK_TEXT, value='이미지').click()

time.sleep(SCROLL_PAUSE_TIME)

# 스크롤바의 위치값
last_height = driver.execute_script('return document.body.scrollHeight')

for i in range(6):
    print('last height: ', last_height)
    # 윈도우 창을 0에서부터 가장 밑(scrollHeight)까지 이동
    driver.execute_script('window.scrollTo(0, document.body.scrollHeight)')
    time.sleep(SCROLL_PAUSE_TIME)
    new_height = driver.execute_script('return document.body.scrollHeight')
    print('new height: ', new_height)
    print('-'*30)
    
# 더 이상 스크롤 될 페이지가 없을 경우 scrollHeight 값의 변화가 없다.
```

## 구글에서 이미지 검색 후 검색 결과 무한 스크롤하기 step.1

```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time

SCROLL_PAUSE_TIME = 1
SEARCH_KEYWORD = 'python'

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=webdriver.ChromeOptions())

driver.get("https://www.google.com")
element = driver.find_element(by=By.CLASS_NAME, value='gLFyf')
element.send_keys(SEARCH_KEYWORD)
element.submit() 

driver.find_element(by=By.LINK_TEXT, value='이미지').click()

time.sleep(SCROLL_PAUSE_TIME)

last_height = driver.execute_script('return document.body.scrollHeight')

# 스크롤 횟수 저장
scroll_cnt = 1

while True:
    driver.execute_script('window.scrollTo(0, document.body.scrollHeight)')
    time.sleep(SCROLL_PAUSE_TIME)
    new_height = driver.execute_script('return document.body.scrollHeight')
    print(f'last height: {last_height}, new_height: {new_height}, scroll_count: {scroll_cnt}')
    
    if last_height != new_height: # 계속해서 new_height 값이 변경된다면...
        last_height = new_height
        scroll_cnt += 1
    else: # 더 이상 new_height 값의 변동이 없다면...
        # CSS selector를 이용한 태그 접근
        ## 해당 element와 상호작용이 불가능해지면 
        ## ElementNotInteractableException 예외 발생
        element = driver.find_element(by=By.CSS_SELECTOR, value='#islmp > div > div > div > div > div.gBPM8 > div.qvfT1 > div.YstHxe > input')
        # select 함수에서도 사용 가능
        element.click()
        print('결과 더보기 클릭')
```

## 구글에서 이미지 검색 후 검색 결과 무한 스크롤하기 step.2

```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time

SCROLL_PAUSE_TIME = 1
SEARCH_KEYWORD = '메이플스토리 팬텀'
IMAGE_EXTRACT_NUM = 10 # 이미지 추출 개수

########### 구글 접속 ###########
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=webdriver.ChromeOptions())
driver.get("https://www.google.com")

########### 검색 ###########
element = driver.find_element(by=By.CLASS_NAME, value='gLFyf')
element.send_keys(SEARCH_KEYWORD)
element.submit() 

########### 이미지 검색결과 페이지 이동 ###########
driver.find_element(by=By.LINK_TEXT, value='이미지').click()
time.sleep(SCROLL_PAUSE_TIME)

########### 이미지 검색결과 페이지 스크롤 ###########
last_height = driver.execute_script('return document.body.scrollHeight')

scroll_cnt = 1

while True:
    driver.execute_script('window.scrollTo(0, document.body.scrollHeight)')
    time.sleep(SCROLL_PAUSE_TIME)
    new_height = driver.execute_script('return document.body.scrollHeight')
#    print(f'last height: {last_height}, new_height: {new_height}, scroll_count: {scroll_cnt}')
    
    if last_height != new_height:
        last_height = new_height
        scroll_cnt += 1
    else:
        element = driver.find_element(by=By.CSS_SELECTOR, value='#islmp > div > div > div > div > div.gBPM8 > div.qvfT1 > div.YstHxe > input')
        try:
            element.click()
            print('결과 더보기 클릭')
        except:
            break

########### 이미지 선택 & 해당 이미지 src 저장 ###########
# css selector 값에 :nth-child(#)은 동일한 태그로 리스트가 있을 경우 리스트 중 특정한 값만 가리킴
# 따라서 전체 리스트를 가져오고자 할 경우에는 :nth-child(#)를 삭제한다.
imgs = driver.find_elements(By.CSS_SELECTOR, '#islrg > div.islrc > div > a.wXeWr.islib.nfEiy')
print('이미지 개수:', len(imgs))

img_src_list = []

img_cnt = 0

for cnt, img in enumerate(imgs, start=1):
    img.send_keys(Keys.ENTER)
    img_a = driver.find_element(By.CSS_SELECTOR, '#Sva75c > div.DyeYj > div > div.dFMRD > div.pxAole > div.tvh9oe.BIB1wf > c-wiz > div.nIWXKc.JgfpDb > div.OUZ5W > div.zjoqD > div.qdnLaf.isv-id.b0vFpe > div > a > img')
    img_src = img_a.get_attribute('src')
    # get_attribute: selenium 속성값 가져오기
    if not img_src.startswith('http'):
        continue
    img_cnt += 1
    print(f'[{img_cnt}]:{img_src}')
    img_src_list.append(img_src)
    
    if img_cnt == IMAGE_EXTRACT_NUM:
        break
```

## 구글에서 이미지 검색 후 검색 결과 무한 스크롤하기 step.3

```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
from bs4 import BeautifulSoup
import requests
import os

SCROLL_PAUSE_TIME = 1
SEARCH_KEYWORD = '김석진'
IMAGE_EXTRACT_NUM = 10 # 이미지 추출 개수

########### 구글 접속 ###########
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=webdriver.ChromeOptions())
driver.get("https://www.google.com")

########### 검색 ###########
element = driver.find_element(by=By.CLASS_NAME, value='gLFyf')
element.send_keys(SEARCH_KEYWORD)
element.submit() 

########### 이미지 검색결과 페이지 이동 ###########
driver.find_element(by=By.LINK_TEXT, value='이미지').click()
time.sleep(SCROLL_PAUSE_TIME)

########### 이미지 검색결과 페이지 스크롤 ###########
last_height = driver.execute_script('return document.body.scrollHeight')

scroll_cnt = 1

while True:
    driver.execute_script('window.scrollTo(0, document.body.scrollHeight)')
    time.sleep(SCROLL_PAUSE_TIME)
    new_height = driver.execute_script('return document.body.scrollHeight')
#    print(f'last height: {last_height}, new_height: {new_height}, scroll_count: {scroll_cnt}')
    
    if last_height != new_height:
        last_height = new_height
        scroll_cnt += 1
    else:
        element = driver.find_element(by=By.CSS_SELECTOR, value='#islmp > div > div > div > div > div.gBPM8 > div.qvfT1 > div.YstHxe > input')
        try:
            element.click()
            print('결과 더보기 클릭')
        except:
            break

########### 이미지 선택 & 해당 이미지 src 저장 ###########
imgs = driver.find_elements(By.CSS_SELECTOR, '#islrg > div.islrc > div > a.wXeWr.islib.nfEiy')
print('이미지 개수:', len(imgs))

img_src_list = []

img_cnt = 0

for cnt, img in enumerate(imgs, start=1):
    img.send_keys(Keys.ENTER)
    img_a = driver.find_element(By.CSS_SELECTOR, '#Sva75c > div.DyeYj > div > div.dFMRD > div.pxAole > div.tvh9oe.BIB1wf > c-wiz > div.nIWXKc.JgfpDb > div.OUZ5W > div.zjoqD > div.qdnLaf.isv-id.b0vFpe > div > a > img')
    img_src = img_a.get_attribute('src')
    if not img_src.startswith('http'):
        continue
    img_cnt += 1
    print(f'[{img_cnt}]:{img_src}')
    img_src_list.append(img_src)
    
    if img_cnt == IMAGE_EXTRACT_NUM:
        break
        
########### 검색 이미지 파일로 저장 ###########

path = 'c:/Temp/'
ts = time.localtime()
now = f'{ts.tm_year}.{ts.tm_mon}.{ts.tm_mday}.{ts.tm_hour}.{ts.tm_min}.{ts.tm_sec}'
directory = SEARCH_KEYWORD+'/'+now+'/'

# change diretory
os.chdir(path)

if not os.path.exists(directory):
    os.makedirs(directory)

file_no = 1
os.chdir(path+directory)
for url in img_src_list:
    extension = url.split('.')[-1]
    ext = '' # 최종적으로 파일의 이름에 붙일 확장자
    if extension in ['jpg','JPG','JPEG','jpeg','png','PNG','gif','GIF']:
        ext = '.'+extension
    else:
        ext = '.jpg'
        
    file_name = str(file_no)+'-'+SEARCH_KEYWORD+ext
    
    response = requests.get(url)
    with open(file_name, 'wb') as f:
        f.write(response.content)
    
    file_no += 1
    
driver.close()
```

# [실습] yes24 사이트에서 파이썬 도서검색하기
- yes24 사이트에서 파이썬 도서 검색 후 평점 9.6 이상인 도서 제목과 가격, 평점 가져오기

```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.select import Select

SEARCH_KEYWORD = '파이썬'

########### 구글 접속 ###########
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=webdriver.ChromeOptions())
driver.get("http://www.yes24.com/main/default.aspx")

########### 검색 ###########
element = driver.find_element(by=By.ID, value='query')
element.send_keys(SEARCH_KEYWORD)
element.submit()

select = Select(driver.find_elements(By.ID, 'stat_gb')[1])
select.select_by_index(3) # 하위태그인 옵션의 위치값; 0부터 시작
select.select_by_value('120') # option 태그의 value 속성 값
select.select_by_visible_text('120개씩 보기') # select 태그의 문자열 값
```

```python
from bs4 import BeautifulSoup
bs = BeautifulSoup(driver.page_source, 'html.parser')
temp = bs.find_all("div", attrs={'class':'item_info'})

rating = 0
book_list = []

for i in temp:
    if i.select_one("#yesSchList > li > div > div.item_info > div.info_row.info_rating > span.rating_grade > em") == None:
        continue
    else:
        rating = float(i.select_one("#yesSchList > li > div > div.item_info > div.info_row.info_rating > span.rating_grade > em").string)
        if rating >= 9.6:
            book_name = i.select_one("#yesSchList > li > div > div.item_info > div.info_row.info_name > a.gd_name").string
            book_price = i.select_one("#yesSchList > li > div > div.item_info > div.info_row.info_price > strong > em").string
            book_info = [book_name, book_price, rating]
            book_list.append(book_info)
        
for i, BOOK in enumerate(book_list, start = 1):
    print(f'{i:2d}번: {BOOK}')
```
