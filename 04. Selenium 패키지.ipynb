```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time

SCROLL_PAUSE_TIME = 1
SEARCH_KEYWORD = 'python'

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=webdriver.ChromeOptions())

driver.get("https://www.google.com")
element = driver.find_element(by=By.CLASS_NAME, value='gLFyf')
element.send_keys(SEARCH_KEYWORD)
element.submit() 

driver.find_element(by=By.LINK_TEXT, value='ì´ë¯¸ì§€').click()

time.sleep(SCROLL_PAUSE_TIME)

last_height = driver.execute_script('return document.body.scrollHeight')

# ìŠ¤í¬ë¡¤ íšŸìˆ˜ ì €ì¥
scroll_cnt = 1

while True:
    driver.execute_script('window.scrollTo(0, document.body.scrollHeight)')
    time.sleep(SCROLL_PAUSE_TIME)
    new_height = driver.execute_script('return document.body.scrollHeight')
    print(f'last height: {last_height}, new_height: {new_height}, scroll_count: {scroll_cnt}')
    
    if last_height != new_height: # ê³„ì†í•´ì„œ new_height ê°’ì´ ë³€ê²½ëœë‹¤ë©´...
        last_height = new_height
        scroll_cnt += 1
    else: # ë” ì´ìƒ new_height ê°’ì˜ ë³€ë™ì´ ì—†ë‹¤ë©´...
        # CSS selectorë¥¼ ì´ìš©í•œ íƒœê·¸ ì ‘ê·¼
        ## í•´ë‹¹ elementì™€ ìƒí˜¸ì‘ìš©ì´ ë¶ˆê°€ëŠ¥í•´ì§€ë©´ 
        ## ElementNotInteractableException ì˜ˆì™¸ ë°œìƒ
        element = driver.find_element(by=By.CSS_SELECTOR, value='#islmp > div > div > div > div > div.gBPM8 > div.qvfT1 > div.YstHxe > input')
        # select í•¨ìˆ˜ì—ì„œë„ ì‚¬ìš© ê°€ëŠ¥
        element.click()
        print('ê²°ê³¼ ë”ë³´ê¸° í´ë¦­')
```

```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
from bs4 import BeautifulSoup
import requests
import os

SCROLL_PAUSE_TIME = 1
SEARCH_KEYWORD = 'ê¹€ì„ì§„'
IMAGE_EXTRACT_NUM = 10 # ì´ë¯¸ì§€ ì¶”ì¶œ ê°œìˆ˜

########### êµ¬ê¸€ ì ‘ì† ###########
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=webdriver.ChromeOptions())
driver.get("https://www.google.com")

########### ê²€ìƒ‰ ###########
element = driver.find_element(by=By.CLASS_NAME, value='gLFyf')
element.send_keys(SEARCH_KEYWORD)
element.submit() 

########### ì´ë¯¸ì§€ ê²€ìƒ‰ê²°ê³¼ í˜ì´ì§€ ì´ë™ ###########
driver.find_element(by=By.LINK_TEXT, value='ì´ë¯¸ì§€').click()
time.sleep(SCROLL_PAUSE_TIME)

########### ì´ë¯¸ì§€ ê²€ìƒ‰ê²°ê³¼ í˜ì´ì§€ ìŠ¤í¬ë¡¤ ###########
last_height = driver.execute_script('return document.body.scrollHeight')

scroll_cnt = 1

while True:
    driver.execute_script('window.scrollTo(0, document.body.scrollHeight)')
    time.sleep(SCROLL_PAUSE_TIME)
    new_height = driver.execute_script('return document.body.scrollHeight')
#    print(f'last height: {last_height}, new_height: {new_height}, scroll_count: {scroll_cnt}')
    
    if last_height != new_height:
        last_height = new_height
        scroll_cnt += 1
    else:
        element = driver.find_element(by=By.CSS_SELECTOR, value='#islmp > div > div > div > div > div.gBPM8 > div.qvfT1 > div.YstHxe > input')
        try:
            element.click()
            print('ê²°ê³¼ ë”ë³´ê¸° í´ë¦­')
        except:
            break

########### ì´ë¯¸ì§€ ì„ íƒ & í•´ë‹¹ ì´ë¯¸ì§€ src ì €ì¥ ###########
imgs = driver.find_elements(By.CSS_SELECTOR, '#islrg > div.islrc > div > a.wXeWr.islib.nfEiy')
print('ì´ë¯¸ì§€ ê°œìˆ˜:', len(imgs))

img_src_list = []

img_cnt = 0

for cnt, img in enumerate(imgs, start=1):
    img.send_keys(Keys.ENTER)
    img_a = driver.find_element(By.CSS_SELECTOR, '#Sva75c > div.DyeYj > div > div.dFMRD > div.pxAole > div.tvh9oe.BIB1wf > c-wiz > div.nIWXKc.JgfpDb > div.OUZ5W > div.zjoqD > div.qdnLaf.isv-id.b0vFpe > div > a > img')
    img_src = img_a.get_attribute('src')
    if not img_src.startswith('http'):
        continue
    img_cnt += 1
    print(f'[{img_cnt}]:{img_src}')
    img_src_list.append(img_src)
    
    if img_cnt == IMAGE_EXTRACT_NUM:
        break
        
########### ê²€ìƒ‰ ì´ë¯¸ì§€ íŒŒì¼ë¡œ ì €ì¥ ###########

path = 'c:/Temp/'
ts = time.localtime()
now = f'{ts.tm_year}.{ts.tm_mon}.{ts.tm_mday}.{ts.tm_hour}.{ts.tm_min}.{ts.tm_sec}'
directory = SEARCH_KEYWORD+'/'+now+'/'

# change diretory
os.chdir(path)

if not os.path.exists(directory):
    os.makedirs(directory)

file_no = 1
os.chdir(path+directory)
for url in img_src_list:
    extension = url.split('.')[-1]
    ext = '' # ìµœì¢…ì ìœ¼ë¡œ íŒŒì¼ì˜ ì´ë¦„ì— ë¶™ì¼ í™•ì¥ì
    if extension in ['jpg','JPG','JPEG','jpeg','png','PNG','gif','GIF']:
        ext = '.'+extension
    else:
        ext = '.jpg'
        
    file_name = str(file_no)+'-'+SEARCH_KEYWORD+ext
    
    response = requests.get(url)
    with open(file_name, 'wb') as f:
        f.write(response.content)
    
    file_no += 1
    
driver.close()
```

```python
from bs4 import BeautifulSoup
bs = BeautifulSoup(driver.page_source, 'html.parser')
temp = bs.find_all("div", attrs={'class':'item_info'})

rating = 0
book_list = []

for i in temp:
    if i.select_one("#yesSchList > li > div > div.item_info > div.info_row.info_rating > span.rating_grade > em") == None:
        continue
    else:
        rating = float(i.select_one("#yesSchList > li > div > div.item_info > div.info_row.info_rating > span.rating_grade > em").string)
        if rating >= 9.6:
            book_name = i.select_one("#yesSchList > li > div > div.item_info > div.info_row.info_name > a.gd_name").string
            book_price = i.select_one("#yesSchList > li > div > div.item_info > div.info_row.info_price > strong > em").string
            book_info = [book_name, book_price, rating]
            book_list.append(book_info)
        
for i, BOOK in enumerate(book_list, start = 1):
    print(f'{i:2d}ë²ˆ: {BOOK}')
```
